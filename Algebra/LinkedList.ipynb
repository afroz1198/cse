### 1. Singly Linked List

```python
class Error(Exception):
    """Base class for exceptions in this module."""
    pass

class EmptyListError(Error):
    """Exception raised when fiddling Empty List.

    Attributes:
        expression -- input expression in which the error occurred
        message -- explanation of the error
    """

    def __init__(self, expression, message):
        self.expression = expression
        self.message = message


class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
        self.length = 0
    def append(self,data):
        new_node= Node(data)
        if not self.head:
            self.head = new_node
            self.length += 1
            return
        temp = self.head
        while(temp.next):
            temp = temp.next
        temp.next = new_node
        self.length += 1      
    def prepend(self,data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.length += 1
            return
        prev_head = self.head
        self.head = new_node
        self.head.next = prev_head
        self.length += 1
    def display(self):
        temp = self.head
        while(temp):
            print(temp.data,end=" ")
            temp= temp.next
        print()
    def insert(self,data,index):
        new_node = Node(data)
        if not self.head:
            self.head=new_node
            self.length += 1
            return
        if index > self.length -1:
            raise IndexError("Index out of bounds,{}(index) > {}".format(index,self.length -1))
            return
        curr_index = 0
        temp = self.head
        while(temp.next and curr_index != index):
            temp = temp.next
            curr_index += 1
        new_node.next = temp.next
        temp.next = new_node
        self.length += 1
    def pop(self):
        if not self.head:
            raise EmptyListError('self.head is None','List is Empty')
            return
        if self.length == 1:
            popped = self.head.data
            self.head = None
            self.length -= 1
            return popped
        temp = self.head
        while(temp.next.next):
            temp = temp.next
        popped = temp.next.data
        temp.next = None
        self.length -= 1
        return popped
    def prepop(self):
        if not self.head:
            raise EmptyListError('self.head is None','List is Empty')
            return
        if self.length == 1:
            popped = self.head
            self.head = None
            self.length -= 1
            return popped
        popped = self.head.data
        self.head = self.head.next
        self.length -= 1
        return popped
def testLinkedList():
    testList = LinkedList()   
    testList.append(4)
    testList.append(5)
    testList.append(6)
    testList.prepend(3)
    testList.prepend(2)
    # insert at middle
    testList.insert(5,2)
    # insert at end
    testList.insert(7,5)
    assert(testList.length == 7)
    # Checking out of bounds
    assert(testList.prepop()==2)
    try:
        testList.insert(11,7)
    except (IndexError):
        pass
    assert(testList.length == 6)
    
    for _ in range(testList.length):
        testList.pop()
    
    try:
        testList.pop()
    except(EmptyListError):
        pass
testLinkedList()
```
